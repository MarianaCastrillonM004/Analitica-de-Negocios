```python
!pip install tensorflow
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#Configurar librerias para la construccion de un modelo neuronal
from keras.models import Sequential #capas para la red neuronal
from keras.layers import Dense #Numero de neuronas por capa
```

##WACCS

```python
nxl="/content/7. BGD and Cost of Debt - Feb 2023_Int.xlsx"
XDB=pd.read_excel(nxl, sheet_name=0, header=[0, 1])

c2022 = XDB.loc[:, (slice(None), 2022)]      # todas las métricas, año = 2022
c2022.columns = c2022.columns.droplevel(1)

XDB = c2022[["ESG Score", "Environmental Pillar Score", "WACC Short Term Debt Cost, (%)" ,"Social Pillar Score", "Governance Pillar Score", "Board Gender Diversity, Percent Score", "Board Size", "Female on Board", "Independent Board Members Score", "Executive Members Gender Diversity, Percent"]]
XDB=XDB.dropna()

#Establecer variable de entrada y de salida
XD=np.array(XDB[["ESG Score", "Environmental Pillar Score","Social Pillar Score", "Governance Pillar Score", "Board Gender Diversity, Percent Score", "Board Size", "Female on Board", "Independent Board Members Score", "Executive Members Gender Diversity, Percent"]])
XDn=XD/np.max(XD, axis=0)
YD=np.array(XDB[ "WACC Short Term Debt Cost, (%)"])
YDn=YD

print("Las variables de entrada son: ",XDn)
print("Las variables de salida son: ",YDn)
```

```python
madaline_sig_WACCS=tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, input_shape=(9,), activation='relu',use_bias=False),
    tf.keras.layers.Dense(1, activation='sigmoid', use_bias=False)
])
madaline_sig_WACCS.load_weights('/content/7. pesos_sigmoid_WACCS.weights.h5')

madaline_sig_WACCS.summary()
madaline_sig_WACCS.compile(optimizer='sgd', loss='mse') #mse: mean square error- quiero que sea cero
                                              #sgd: solve descend gradient
history=madaline_sig_WACCS.fit(XDn,YDn, epochs=250)
ydp=madaline_sig_WACCS.predict(XDn)



#Scatter plot
plt.figure()
plt.scatter(YDn, ydp, label='Grafico de correlacion')
plt.grid()
plt.show()

#Figura entre curvas
plt.figure()
plt.plot(YDn, 'r', label='Curva real')
plt.plot(ydp, 'b', label='Curva predicha')
plt.grid()
plt.legend()
plt.show()

#Determinar la correlacion
df=pd.DataFrame(np.column_stack((YDn,ydp)))
print("la correlacion de los datos es:", df.corr())

#Efectos independientes - Importancia de las variables de entrada

WC=madaline_sig_WACCS.get_weights()
W= WC[0]; c=WC[1] #la capa 0 - 4 a 10; capa 1 - 10 -1 salida: conexiones nerviosas (1 regresiones)
a=W@c
print("Los efectos independientes son: \n", a)

#Cuando el efecto independiente es positivo, la prob de default aumenta
#Cuando el efecto independiente es negativo, la prob de default disminuye
```

```python
#para evaluar Empresa
XDo=[59.29,16.17,83.88,42.66,72.01,6,16.67,3.36,26.87],
Xscn=XDo/np.max(XD, axis=0)
ydpsc=madaline_sig_WACCS.predict(Xscn)
print('El score estimado es:', ydpsc)
```

#Autoencoders


```python
#Autoencoders
#Misma entrada - Misma salida

nxl="/content/7. BGD and Cost of Debt - Feb 2023_Int.xlsx"

XDB=pd.read_excel(nxl, sheet_name=0, header=[0, 1])

c2022 = XDB.loc[:, (slice(None), 2022)]      # todas las métricas, año = 2022
c2022.columns = c2022.columns.droplevel(1)

XDB = c2022[["ESG Score", "Environmental Pillar Score" ,"Social Pillar Score", "Governance Pillar Score", "Board Gender Diversity, Percent Score", "Board Size", "Female on Board", "Independent Board Members Score", "Executive Members Gender Diversity, Percent", "WACC Short Term Debt Cost, (%)" ,"WACC Long Term Debt Cost, (%)"]]
XDB=XDB.dropna()

XD=np.array(XDB)         #misma entrada
XDn=XD/np.max(XD,axis=0) #normalización variables
ydna=np.array(XDn)       #misma salida

madaline_sig=tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, input_shape=(11,), activation='relu',use_bias=False),
    tf.keras.layers.Dense(11, activation='sigmoid', use_bias=False)])

# Se cargan los pesos entrenados previamente para los modelos Adaline y Madaline

madaline_sig.load_weights('/content/7. pesos_autoencoder_BGD.weights.h5')

madaline_sig.summary()


madaline_sig.compile(optimizer='sgd', loss='mse') #sgd: solve descend gradient
history=madaline_sig.fit(XDn,ydna, epochs=250) #recorre la tabla 100 veces
ydap=madaline_sig.predict(XDn) #ofrece una predicción

#Las correlaciones son las siguientes:
df=pd.DataFrame(np.column_stack((XDn,ydap)))
print("la correlacion de los datos es: \n", df.corr())
```
